```python
import pandas as pd
import numpy as np
```

```python
data ={
    'Name':['Jai', 'Princi', 'Gaurav', 'Anuj'],
    'Age':[27, 24, 22, 32],
    'Gender' : ['Male', 'Female', 'Male', 'Male'],
    'Marks' : [90,91,43,75]
}
data
data = pd.DataFrame(data)
data
```

```python
cat =[]
con =[]

for i in data.columns:
  if data[i].dtype == "object":
    cat.append(i)
  else:
    con.append(i)
cat
```

```python
con
```

```python
data.drop(['Age'] , axis=1 )
```

```python
c=sum=avg=0

for ele in data['Marks']:
  if str(ele).isnumeric():
    c+=1
    sum+=ele
if c>0:
  avg=sum/c

data.replace(to_replace='Nan' ,value=avg)
```

```python
data.drop([2], axis=0)
```

```python
data['Gender'] = data['Gender'].map({'Male':0,'Female':1}).fillna(-1).astype(int)

print(data)
```

```python
data = data[data['Age']>30]
data
```
